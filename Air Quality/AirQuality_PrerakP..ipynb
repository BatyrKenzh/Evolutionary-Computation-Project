```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import mean_squared_error
import itertools
import zipfile
import requests
import os
import matplotlib.pyplot as plt

# Download and extract the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00360/AirQualityUCI.zip"
zip_file = "AirQualityUCI.zip"
csv_file = "AirQualityUCI.csv"

if not os.path.exists(csv_file):
    # Download the zip file
    response = requests.get(url)
    with open(zip_file, 'wb') as f:
        f.write(response.content)

    # Extract the CSV file
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall()

# Load the dataset
data = pd.read_csv(csv_file, sep=';', decimal=',', na_values=-200.0)

# Preprocess the dataset
def preprocess_data(data):
    # Keep only numeric columns and drop irrelevant columns
    data = data.iloc[:, :-2]

    # Drop rows with missing values
    data = data.dropna()

    # Rename columns for clarity
    data.columns = ['Date', 'Time', 'CO(GT)', 'PT08.S1(CO)', 'NMHC(GT)', 'C6H6(GT)',
                    'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)',
                    'PT08.S5(O3)', 'T', 'RH', 'AH']

    # Convert Date and Time to a single datetime column
    data['Datetime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'], format='%d/%m/%Y %H.%M.%S')
    data = data.drop(['Date', 'Time'], axis=1)

    # Set the target variable (NOx)
    target = 'NOx(GT)'

    # Normalize features
    scaler = MinMaxScaler()
    data.iloc[:, :-1] = scaler.fit_transform(data.iloc[:, :-1])

    return data, target, scaler

data, target, scaler = preprocess_data(data)

# Create lagged features for time series
def create_time_series(data, target, lookback=10):
    X, y = [], []
    for i in range(lookback, len(data)):
        X.append(data.iloc[i-lookback:i, :-1].values)  # Features (lagged data)
        y.append(data.iloc[i][target])  # Target (current NOx value)
    return np.array(X), np.array(y)

lookback = 10
X, y = create_time_series(data, target, lookback)

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Define hyperparameter search space
hyperparameter_space = {
    'neurons': [32],
    'dropout': [0.1, 0.2, 0.3],
    'learning_rate': [0.001, 0.005, 0.01],
    'batch_size': [16],
    'epochs': [50],
    'model_type': ['LSTM', 'GRU']  # LSTM or GRU
}

# Build the model dynamically
def build_model(neurons, dropout, learning_rate, model_type):
    model = Sequential()

    if model_type == 'LSTM':
        model.add(LSTM(neurons, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=False))
    elif model_type == 'GRU':
        model.add(GRU(neurons, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=False))

    model.add(Dropout(dropout))
    model.add(Dense(1))
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate), loss='mse', metrics=['mae'])
    return model

# Perform hyperparameter optimization
def hyperparameter_optimization():
    best_model = None
    best_params = None
    best_rmse = float('inf')
    best_history = None
    # Grid search through hyperparameters
    for neurons, dropout, learning_rate, batch_size, epochs, model_type in itertools.product(
        hyperparameter_space['neurons'],
        hyperparameter_space['dropout'],
        hyperparameter_space['learning_rate'],
        hyperparameter_space['batch_size'],
        hyperparameter_space['epochs'],
        hyperparameter_space['model_type']
    ):
        print(f"Testing config: neurons={neurons}, dropout={dropout}, lr={learning_rate}, bs={batch_size}, ep={epochs}, type={model_type}")

        model = build_model(neurons, dropout, learning_rate, model_type)
        early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

        history = model.fit(
            X_train, y_train,
            validation_data=(X_test, y_test),
            batch_size=batch_size,
            epochs=epochs,
            callbacks=[early_stopping],
            verbose=0
        )

        y_pred = model.predict(X_test)
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        if rmse < best_rmse:
            best_rmse = rmse
            best_model = model
            best_params = {
                'neurons': neurons,
                'dropout': dropout,
                'learning_rate': learning_rate,
                'batch_size': batch_size,
                'epochs': epochs,
                'model_type': model_type
            }
            best_history = history
    print("Best Hyperparameters:", best_params)
    print("Best RMSE:", best_rmse)
    return best_model, best_params, best_rmse, best_history

# Run hyperparameter optimization
best_model, best_params, best_rmse, best_history = hyperparameter_optimization()

# Evaluate the best model
y_pred = best_model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Best Model RMSE:", rmse)

# Plot training history for the best model
def plot_training_history(history):
    # Extract metrics from the history
    train_loss = history.history['loss']
    val_loss = history.history['val_loss']
    train_mae = history.history['mae']
    val_mae = history.history.get('val_mae', [])

    # Plot training and validation loss
    plt.figure(figsize=(12, 6))
    plt.plot(train_loss, label='Training Loss', color='blue')
    plt.plot(val_loss, label='Validation Loss', color='orange')
    plt.title('Loss During Training')
    plt.xlabel('Epochs')
    plt.ylabel('Loss (MSE)')
    plt.legend()
    plt.grid()
    plt.show()

    # Plot training and validation MAE
    if val_mae:
        plt.figure(figsize=(12, 6))
        plt.plot(train_mae, label='Training MAE', color='green')
        plt.plot(val_mae, label='Validation MAE', color='red')
        plt.title('MAE During Training')
        plt.xlabel('Epochs')
        plt.ylabel('MAE')
        plt.legend()
        plt.grid()
        plt.show()

# Plot predicted vs actual values
def plot_predictions(y_test, y_pred):
    plt.figure(figsize=(12, 6))
    plt.plot(y_test, label='Actual Values', color='blue', alpha=0.6)
    plt.plot(y_pred, label='Predicted Values', color='orange', alpha=0.6)
    plt.title('Predicted vs Actual Values')
    plt.xlabel('Samples')
    plt.ylabel('NOx(GT)')
    plt.legend()
    plt.grid()
    plt.show()

# Run training and generate plots
print("Starting training history plot...")
plot_training_history(best_history)

print("Plotting predictions vs actual values...")
plot_predictions(y_test, y_pred)
```
